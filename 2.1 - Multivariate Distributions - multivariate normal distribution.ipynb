Lets $\textbf{z}=\left(Z_1,...,Z_p\right)$ a random vector with dimension p of independent random variables such that each entrie $Z_i \sim  N(0,1)$ $i=1,..,p$ and consider the following invertible transformation $T:\mathbb{R}^p \rightarrow \mathbb{R}^P$ given by the linear combination of the elements of the random vector followed by translation:

$$
\begin{align*}
X_{1}   & = \mu_1 + a_{1 1} Z_{1} + \cdots + a_{1 p} Z_{p}   \\
\vdots  & = \vdots \\
X_{p}   & = \mu_p + a_{p 1} Z_{1} + \cdots + a_{p p} Z_{p} \\
\\
\left[ \begin{array}{ccccc}
X_{1} \\
\vdots  \\
X_{p} \end{array} \right]
& = 
\left[ \begin{array}{ccccc}
\mu_{1} \\
\vdots  \\
\mu_{p} \end{array} \right]
+
\left[ \begin{array}{ccccc}
a_{1 1} & \cdots & a_{1 p} \\
\vdots  & \ddots & \vdots \\
a_{p 1} & \cdots & a_{p p} \end{array} \right]
\left[ \begin{array}{ccccc}
Z_{1} \\
\vdots  \\
Z_{p} \end{array} \right]\\
\\
\textbf{x} & = \mu +\textbf{Az} \\
\textbf{x} & = T \left[ \textbf{z} \right] \\
\end{align*}
$$

Hence it follows that for $T$ to be invertible then A must be invertible. A natural question is to find the joint distribution of the new random vector $\textbf{X}$, so in order to fin that density function it's necesary apply the change of variable theorem for jointly distributions.

$$
\begin{align*}
f_{X_1,...,X_p}(x_1,...,x_p)&=f_{Z_1,...,Z_p}(T_1^{-1}(X),...,T_p^{-1}(X))\left| \frac{\partial T^{-1}}{\partial X} \right|\\
\\
\textbf{x} & = \mu + \textbf{Az} \\
\textbf{x} - \mu  & = \textbf{Az} \\
\textbf{A}^{-1}\left[\textbf{x} - \mu \right]  & = \textbf{z} ..... (1)\\
T^{-1}\left[\textbf{X}\right]  & = \textbf{z} \\
\\
\implies \frac{\partial T^{-1}\left[\textbf{x}\right]}{\partial \textbf{x}} & = \frac{\partial}{\partial \textbf{x}} \left( \textbf{A}^{-1}\left[\textbf{x} - \mu \right]  \right)\\
& = \frac{\partial}{\partial \textbf{x}} \textbf{A}^{-1} \textbf{x} -  \frac{\partial}{\partial \textbf{x}}\textbf{A}^{-1}\mu\\
& = \frac{\partial}{\partial \textbf{x}} \textbf{A}^{-1} \textbf{x} -  \textbf{0}\\
\frac{\partial T^{-1}\left[\textbf{x}\right]}{\partial \textbf{x}} & = \textbf{A}^{-1} \\
\left| \frac{\partial T^{-1}\left[\textbf{x}\right]}{\partial \textbf{x}} \right |& = \left| \textbf{A}^{-1} \right| ..... (2)\\
\end{align*}
$$

On the other hand since $\mathbf{x}$ it's a random vector of independent random variables then the jointly distribution it's given by the product of the marginal distributions.

$$
\begin{align*}
f_{Z_1,...,Z_p}(z_1,...,z_p) & = f_{Z_1}(z_1) \cdots f_{Z_p}(z_p) \\
& = \left[ \frac{1}{\sqrt{2 \pi}} e^{-\frac{1}{2}z_1^2} \right] \cdots  \left[ \frac{1}{\sqrt{2 \pi}} e^{-\frac{1}{2}z_p^2} \right]\\
& =  \left[\frac{1}{\sqrt{2 \pi}}\right]^p exp\left( \frac{1}{2} \sum_{k=0}^p z_i^2\right)\\
& =  \left[\frac{1}{\sqrt{2 \pi}}\right]^p exp\left( \frac{1}{2} \textbf{z' z}\right)\\
\\
\implies f_{Z_1,...,Z_p}(T_1^{-1}(X),...,T_p^{-1}(X))\left| \frac{\partial T^{-1}}{\partial X} \right| &= \left[\frac{1}{\sqrt{2 \pi}}\right]^p exp\left( \frac{1}{2} \textbf{z' z}\right) \left| \textbf{A}^{-1} \right| ..... (3)\\
\end{align*}
$$

From the equation (1) its possible obtain $\mathbf{z'z}$ in termns of $\mathbf{x}$

$$
\begin{align*}
\textbf{z}  & = \textbf{A}^{-1}\left[\textbf{x} - \mu \right]\\
\textbf{z' z}  & = \left[\textbf{A}^{-1}\left[\textbf{x}-\mu \right]\right]' \textbf{A}^{-1}\left[\textbf{x} - \mu \right]\\
& = \left[\textbf{x}-\mu \right]' \left[\textbf{A}^{-1}\right]' \textbf{A}^{-1}\left[\textbf{x} - \mu \right]\\
& = \left[\textbf{x}-\mu \right]' \left[\textbf{A A'}\right]^{-1}\left[\textbf{x} - \mu \right]\\
\textbf{z' z} & = \left[\textbf{x}-\mu \right]' \Sigma^{-1} \left[\textbf{x} - \mu \right] ..... (4)\\
\end{align*}
$$

Based on the fact that A it's invertible then  $\mathbf{AA'}=\Sigma$ it's positive-definite symmetric matrix, then:

$$
\left| A \right|=\sqrt{\left|A\right|\left|A\right|}=\sqrt{\left|AA'\right|}=\sqrt{\left|\Sigma\right|}=\Sigma^{\frac{1}{2}}\\
\implies \left| A \right|^{-1}=\left| \Sigma \right|^{-\frac{1}{2}}.....(5)
$$

Replacing (4) and (5) in (3) the density of this transformation $\mathbf{T}$ is obtained:

$$
f_{X_1,...,X_p}(x_1,...,x_p) = \frac{\left| \Sigma \right|^{-\frac{1}{2}}}{\left[\sqrt{2 \pi}\right]^p} exp\left( \frac{1}{2} \left[\textbf{x}-\mu \right]' \Sigma^{-1} \left[\textbf{x} - \mu \right] \right)
$$

This multivariate distribution it's called the **multivariate normal distribution**. Note that the density function of the random vector is completely defined by the parameters $\mu$ and $\Sigma$, where $\Sigma$ should be symmetric with non determinant zero. **Notation** We denote that the random vector $\textbf{x}$ distributes multivariate normal with parameters $\mu$ and $\Sigma$ as:
$$
\textbf{x} \sim N\left(\mu , \Sigma \right) 
$$
When $\mu=\textbf{0}$ and $\Sigma=\textbf{T}$ then it is said that x distributes as a normal multivariate standard.

#### Bivariated normal dstribution
___

We call the above joint distribution for the vector $\textbf{x}=(X_1,X_2)$ the bivariated normal distribution when:

$$
\textbf{x} \sim N\left(
\mu=\left[ \begin{array}{ccccc}
X_{1} \\
X_{2} \end{array} \right]
,
\Sigma=\left[ \begin{array}{ccccc}
\sigma_{11} & \sigma_{12}\\
\sigma_{21} & \sigma_{22} \end{array} \right]
\right) 
$$

with density function

$$
\begin{align*}
f_{X_1,X_2}(x_1,x_2) & = \frac{\left| 
\begin{array}{ccccc}
\sigma_{11} & \sigma_{12}\\
\sigma_{21} & \sigma_{22} \end{array}
\right|^{-\frac{1}{2}}}{\left[\sqrt{2 \pi}\right]^2} 
exp\left( \frac{1}{2} 
\left[x_1-\mu_1 x_2-\mu_2 \right]
\left[\begin{array}{ccccc}
\sigma_{11} & \sigma_{12}\\
\sigma_{21} & \sigma_{22} \end{array} \right]^{-1} 
\left[ \begin{array}{ccccc}
x_1-\mu_1 \\
x_2-\mu_2 \end{array} \right]
\right)\\
& = \frac{1}{2 \pi \sqrt{\sigma_{11}\sigma_{22}-\sigma_{21}\sigma_{12}}} 
exp\left( \frac{1}{2\left[\sigma_{11}\sigma_{22}-\sigma_{21}\sigma_{12}\right]}
\left[x_1-\mu_1 x_2-\mu_2 \right]
\left[\begin{array}{ccccc}
\sigma_{22} & -\sigma_{12}\\
-\sigma_{21} & \sigma_{11} \end{array} \right]
\left[ \begin{array}{ccccc}
x_1-\mu_1 \\
x_2-\mu_2 \end{array} \right]
\right)\\
\end{align*}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D

mu=[0,0]
Sigma = [[7,1],[1,7]]

x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)
X, Y = np.meshgrid(x,y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y
rv = multivariate_normal(mu,Sigma)

fig = plt.figure(figsize=(20,10))
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, rv.pdf(pos),cmap='viridis',linewidth=0)
ax.set_xlabel('$X_1$ axis')
ax.set_ylabel('$X_2$ axis')
ax.set_zlabel('$f_{X_1,X_2}(x_1,x_2)$')
ax.view_init(50,10)
plt.show()
```

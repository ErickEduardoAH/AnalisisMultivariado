Lets $\textbf{z}=\left(Z_1,...,Z_p\right)$ a random vector with dimension p of independent random variables such that each entrie $Z_i \sim  N(0,1)$ $i=1,..,p$ and consider the following invertible transformation $T:\mathbb{R}^p \rightarrow \mathbb{R}^P$ given by the linear combination of the elements of the random vector followed by translation:

$$
\begin{align*}
X_{1}   & = \mu_1 + a_{1 1} Z_{1} + \cdots + a_{1 p} Z_{p}   \\
\vdots  & = \vdots \\
X_{p}   & = \mu_p + a_{p 1} Z_{1} + \cdots + a_{p p} Z_{p} \\
\\
\left[ \begin{array}{ccccc}
X_{1} \\
\vdots  \\
X_{p} \end{array} \right]
& = 
\left[ \begin{array}{ccccc}
\mu_{1} \\
\vdots  \\
\mu_{p} \end{array} \right]
+
\left[ \begin{array}{ccccc}
a_{1 1} & \cdots & a_{1 p} \\
\vdots  & \ddots & \vdots \\
a_{p 1} & \cdots & a_{p p} \end{array} \right]
\left[ \begin{array}{ccccc}
Z_{1} \\
\vdots  \\
Z_{p} \end{array} \right]\\
\\
\textbf{x} & = \mu +\textbf{Az} \\
\textbf{x} & = T \left[ \textbf{z} \right] \\
\end{align*}
$$

Hence it follows that for $T$ to be invertible then A must be invertible. A natural question is to find the joint distribution of the new random vector $\textbf{X}$, so in order to fin that density function it's necesary apply the change of variable theorem for jointly distributions.

$$
\begin{align*}
f_{X_1,...,X_p}(x_1,...,x_p)&=f_{Z_1,...,Z_p}(T_1^{-1}(X),...,T_p^{-1}(X))\left| \frac{\partial T^{-1}}{\partial X} \right|\\
\\
\textbf{x} & = \mu + \textbf{Az} \\
\textbf{x} - \mu  & = \textbf{Az} \\
\textbf{A}^{-1}\left[\textbf{x} - \mu \right]  & = \textbf{z} ..... (1)\\
T^{-1}\left[\textbf{X}\right]  & = \textbf{z} \\
\\
\implies \frac{\partial T^{-1}\left[\textbf{x}\right]}{\partial \textbf{x}} & = \frac{\partial}{\partial \textbf{x}} \left( \textbf{A}^{-1}\left[\textbf{x} - \mu \right]  \right)\\
& = \frac{\partial}{\partial \textbf{x}} \textbf{A}^{-1} \textbf{x} -  \frac{\partial}{\partial \textbf{x}}\textbf{A}^{-1}\mu\\
& = \frac{\partial}{\partial \textbf{x}} \textbf{A}^{-1} \textbf{x} -  \textbf{0}\\
\frac{\partial T^{-1}\left[\textbf{x}\right]}{\partial \textbf{x}} & = \textbf{A}^{-1} \\
\left| \frac{\partial T^{-1}\left[\textbf{x}\right]}{\partial \textbf{x}} \right |& = \left| \textbf{A}^{-1} \right| ..... (2)\\
\end{align*}
$$

On the other hand since $\mathbf{x}$ it's a random vector of independent random variables then the jointly distribution it's given by the product of the marginal distributions.

$$
\begin{align*}
f_{Z_1,...,Z_p}(z_1,...,z_p) & = f_{Z_1}(z_1) \cdots f_{Z_p}(z_p) \\
& = \left[ \frac{1}{\sqrt{2 \pi}} e^{-\frac{1}{2}z_1^2} \right] \cdots  \left[ \frac{1}{\sqrt{2 \pi}} e^{-\frac{1}{2}z_p^2} \right]\\
& =  \left[\frac{1}{\sqrt{2 \pi}}\right]^p exp\left( \frac{1}{2} \sum_{k=0}^p z_i^2\right)\\
& =  \left[\frac{1}{\sqrt{2 \pi}}\right]^p exp\left( \frac{1}{2} \textbf{z' z}\right)\\
\\
\implies f_{Z_1,...,Z_p}(T_1^{-1}(X),...,T_p^{-1}(X))\left| \frac{\partial T^{-1}}{\partial X} \right| &= \left[\frac{1}{\sqrt{2 \pi}}\right]^p exp\left( \frac{1}{2} \textbf{z' z}\right) \left| \textbf{A}^{-1} \right| ..... (3)\\
\end{align*}
$$

From the equation (1) its possible obtain $\mathbf{z'z}$ in termns of $\mathbf{x}$

$$
\begin{align*}
\textbf{z}  & = \textbf{A}^{-1}\left[\textbf{x} - \mu \right]\\
\textbf{z' z}  & = \left[\textbf{A}^{-1}\left[\textbf{x}-\mu \right]\right]' \textbf{A}^{-1}\left[\textbf{x} - \mu \right]\\
& = \left[\textbf{x}-\mu \right]' \left[\textbf{A}^{-1}\right]' \textbf{A}^{-1}\left[\textbf{x} - \mu \right]\\
& = \left[\textbf{x}-\mu \right]' \left[\textbf{A A'}\right]^{-1}\left[\textbf{x} - \mu \right]\\
\textbf{z' z} & = \left[\textbf{x}-\mu \right]' \Sigma^{-1} \left[\textbf{x} - \mu \right] ..... (4)\\
\end{align*}
$$

Based on the fact that A it's invertible then  $\mathbf{AA'}=\Sigma$ it's semipositive-definite symmetric matrix, then:

$$
\left| A \right|=\sqrt{\left|A\right|\left|A\right|}=\sqrt{\left|AA'\right|}=\sqrt{\left|\Sigma\right|}=\Sigma^{\frac{1}{2}}\\
\implies \left| A \right|^{-1}=\left| \Sigma \right|^{-\frac{1}{2}}.....(5)
$$

Replacing (4) and (5) in (3) the density of this transformation $\mathbf{T}$ is obtained:

$$
f_{X_1,...,X_p}(x_1,...,x_p) = \frac{\left| \Sigma \right|^{-\frac{1}{2}}}{\left[\sqrt{2 \pi}\right]^p} exp\left( \frac{1}{2} \left[\textbf{x}-\mu \right]' \Sigma^{-1} \left[\textbf{x} - \mu \right] \right)
$$

This multivariate distribution it's called the **multivariate normal distribution**. Note that the density function of the random vector is completely defined by the parameters $\mu$ and $\Sigma$, where $\Sigma$ should be symmetric with non determinant zero. 

**Notation** We denote that the random vector $\textbf{x}$ distributes multivariate normal with parameters $\mu \in \mathbb{R}^p$ and semipositive definite $\Sigma \in \mathbb{R}^p \times \mathbb{R}^p$ as:
$$
\textbf{x} \sim N_p\left(\mu , \Sigma \right) 
$$
When $\mu=\textbf{0}$ and $\Sigma=\textbf{I}$ then it is said that $\mathbf{x}$ distributes as a normal multivariate standard.

The argument $\left[\textbf{x}-\mu \right]' \Sigma^{-1} \left[\textbf{x} - \mu \right]$ it's a squared weighted distance, where $\Sigma = \mathbf{I}$ then $\left[\textbf{x}-\mu \right]' \Sigma^{-1} \left[\textbf{x} - \mu \right] = \left|\left|\mathbf{x}-\mu\right|\right|^2$.

#### Properties

If $\mathbf{x} \in \mathbb{R}^p, \mathbf{x} \sim N\left(\mu , \Sigma \right)$ and
1. $\mathbf{A} \in \mathbb{R}^p \times \mathbb{R}^q$ ma atrix of constants then $\mathbf{x_{1 \times p}A_{p \times q}} \sim \mathbf{N_q(\mu A, A' \Sigma A)}$.
1. $\mathbf{\Sigma=\Sigma^{1/2} \Sigma^{1/2}}$ where $\Sigma^{1/2}\Sigma^{1/2}$ is the decomposition in **Cholesky factorization** of $\mathbf{\Sigma}$ then $\mathbf{z}=\mathbf{\left( x -\mu \right)\Sigma^{-1/2}}\sim N_p\left(\mathbf{0}, \mathbf{I} \right)$

#### Bivariated normal distribution
___

We call the above joint distribution for the vector $\textbf{x}=(X_1,X_2)$ the bivariated normal distribution when:

$$
\textbf{x} \sim N\left(
\mu=\left[ \begin{array}{ccccc}
X_{1} \\
X_{2} \end{array} \right]
,
\Sigma=\left[ \begin{array}{ccccc}
\sigma_{11} & \sigma_{12}\\
\sigma_{21} & \sigma_{22} \end{array} \right]
\right) 
$$

with density function

$$
\begin{align*}
f_{X_1,X_2}(x_1,x_2) & = \frac{\left| 
\begin{array}{ccccc}
\sigma_{11} & \sigma_{12}\\
\sigma_{21} & \sigma_{22} \end{array}
\right|^{-\frac{1}{2}}}{\left[\sqrt{2 \pi}\right]^2} 
exp\left( \frac{1}{2} 
\left[x_1-\mu_1 x_2-\mu_2 \right]
\left[\begin{array}{ccccc}
\sigma_{11} & \sigma_{12}\\
\sigma_{21} & \sigma_{22} \end{array} \right]^{-1} 
\left[ \begin{array}{ccccc}
x_1-\mu_1 \\
x_2-\mu_2 \end{array} \right]
\right).......(a)\\
\end{align*}
$$

This density can be expressed in terms of quadratic equations

$$
\begin{align*}
\left[x_1-\mu_1 x_2-\mu_2 \right]
\left[\begin{array}{ccccc}
\sigma_{11} & \sigma_{12}\\
\sigma_{21} & \sigma_{22} \end{array} \right]^{-1} 
\left[ \begin{array}{ccccc}
x_1-\mu_1 \\
x_2-\mu_2 \end{array} \right]
&=
\frac{1}{\sigma_{11}\sigma_{22}-\sigma_{12}\sigma_{21}}
\left[x_1-\mu_1 x_2-\mu_2 \right]
\left[\begin{array}{ccccc}
\sigma_{22} & -\sigma_{12}\\
-\sigma_{21} & \sigma_{11} \end{array} \right]
\left[ \begin{array}{ccccc}
x_1-\mu_1 \\
x_2-\mu_2 \end{array} \right] \\
&=
\frac{1}{\sigma_{1}^2 \sigma_{2}^2-\sigma_{12}^2}
\left[x_1-\mu_1 x_2-\mu_2 \right]
\left[\begin{array}{ccccc}
\sigma_{2}^2 & -\sigma_{12}\\
-\sigma_{21} & \sigma_{1}^2 \end{array} \right]
\left[ \begin{array}{ccccc}
x_1-\mu_1 \\
x_2-\mu_2 \end{array} \right] \\
\end{align*}
$$

Since $\rho=\rho_{12} = \rho_{21} = \frac{\sigma_{12}}{\sigma_1 \sigma_2} \implies \rho^2\sigma_1^2 \sigma_2^2 = \sigma_{12}^2$ then $\sigma_{1}^2 \sigma_{2}^2-\sigma_{12}^2 = \sigma_{1}^2\sigma_2^2 (1-\rho^2)....(b)$ and considerng the following change  of variables $x_1-\mu_1=x_1$ and $x_2-\mu_2=z_2$ then:

$$
\begin{align*}
\frac{1}{\sigma_{1}^2 \sigma_{2}^2-\sigma_{12}^2}
\left[x_1-\mu_1 x_2-\mu_2 \right]
\left[\begin{array}{ccccc}
\sigma_{2}^2 & -\sigma_{12}\\
-\sigma_{21} & \sigma_{1}^2 \end{array} \right]
\left[ \begin{array}{ccccc}
x_1-\mu_1 \\
x_2-\mu_2 \end{array} \right]
&=
\frac{1}{\sigma_{1}^2\sigma_2^2 (1-\rho^2)}
\left[\begin{array}{ccccc} z_1 & z_2 \end{array} \right]
\left[\begin{array}{ccccc}
\sigma_{2}^2 & -\sigma_{12}\\
-\sigma_{21} & \sigma_{1}^2 \end{array} \right]
\left[ \begin{array}{ccccc}
z_1 \\
z_2 \end{array} \right] \\
& = \frac{1}{\sigma_{1}^2\sigma_2^2 (1-\rho^2)}
\left[z_1^2\sigma_2^2+z_2^2\sigma_{1}-2z_1z_2\sigma_{12}^2\right] \\
& = \frac{1}{(1-\rho^2)} \left[
\frac{z_1^2 \sigma_2^2}{\sigma_{1}^2\sigma_2^2}+\frac{z_2^2\sigma_{1}}{\sigma_{1}^2\sigma_2^2} -\frac{2z_1z_2\sigma_1\sigma_2\rho}{\sigma_{1}^2\sigma_2^2} 
\right]\\
& = \frac{1}{(1-\rho^2)} \left[
\left(\frac{z_1}{\sigma_{1}}\right)^2+\left(\frac{z_2}{\sigma_2}\right)^2 - 2\rho\frac{z_1 z_2}{\sigma_1 \sigma_2} 
\right]\\
\frac{1}{\sigma_{1}^2 \sigma_{2}^2-\sigma_{12}^2}
\left[x_1-\mu_1 x_2-\mu_2 \right]
\left[\begin{array}{ccccc}
\sigma_{2}^2 & -\sigma_{12}\\
-\sigma_{21} & \sigma_{1}^2 \end{array} \right]
\left[ \begin{array}{ccccc}
x_1-\mu_1 \\
x_2-\mu_2 \end{array} \right]
& = \frac{1}{(1-\rho^2)} \left[
\left(\frac{x_1-\mu_1}{\sigma_{1}}\right)^2+\left(\frac{x_2-\mu_2}{\sigma_2}\right)^2 - 2\rho\frac{(x_1-\mu_1)(x_2-\mu_2)}{\sigma_1 \sigma_2} 
\right]....(c)\\
\end{align*}
$$

Replacing $(b)$ and $(c)$ in $(a)$ then the density is expressed as:

$$
\begin{align*}
f_{X_1,X_2}(x_1,x_2) & = \frac{1}{2 \pi (1-\rho^2)} exp\left( 
\frac{1}{2 (1-\rho^2)}
\left[ \left(\frac{x_1-\mu_1}{\sigma_{11}}\right)^2+
\left(\frac{x_2-\mu_2}{\sigma_{22}}\right)^2-
2\rho \frac{(x_1-\mu_1)(x_2-\mu_2)}{\sigma_1 \sigma_2}
\right]
\right)\\
\end{align*}
$$

$$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D

mu=[0,0]
sigma = [[5,5],[5,10]]

x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)
X, Y = np.meshgrid(x,y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y
rv = multivariate_normal(mu,sigma)
fig = plt.figure(figsize=(20,10))
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, rv.pdf(pos),cmap='viridis',linewidth=0)
ax.set_xlabel('$X_1$ axis')
ax.set_ylabel('$X_2$ axis')
ax.set_zlabel('$f_{X_1,X_2}(x_1,x_2)$')
ax.view_init(60,10)
plt.show()
```

# Spectral decomposition of covariance matrix
___

Let's $\mathbf{\Sigma}_{p \times p}$ a covariance matrix, this matrix it's orthogonally diagonalizable, i.e. 
$$\mathbf{\Sigma}=\mathbf{\Gamma \Lambda \Gamma'}$$
Were:
* $\mathbf{\Lambda}$ it's a diagonal matrix formed by the eigenvalues (ordered reals) of $\mathbf{\Sigma}$.
* $\mathbf{\Gamma}$ it's an ortogonal matrix formed by the ordered eigenvectors of $\mathbf{\Sigma}$.

**Proof:**

Since $\mathbf{\Sigma}$ it's a real symmetric matrix with non-negative values by definition, then for some vector $\mathbf{\gamma} \neq 0 \in \mathbb{C}^p$ and some $\lambda \in \mathbb{C}$, an eigenvalue of $\mathbf{\Sigma}$ satisfy $\mathbf{\Sigma \gamma} = \lambda \mathbf{\gamma}$.

$$ \mathbf{\bar{\gamma}'\Sigma \gamma}\in \mathbb{C}\implies \mathbf{\bar{\gamma}'\Sigma \gamma}
=\left(\mathbf{\bar{\gamma}'\Sigma e}\right)'
=\mathbf{\gamma'\Sigma' \bar{\gamma}}
=\mathbf{\gamma'\Sigma \bar{\gamma}}
=\mathbf{\gamma'\overline{\Sigma \gamma}}
=\mathbf{\overline{\overline{\gamma}' \Sigma \gamma}}
$$

So $\mathbf{\bar{\gamma}'\Sigma \gamma}$ its equal to its conjugated, so it's complex part are null $\implies \mathbf{\bar{\gamma}'\Sigma \gamma} \in \mathbb{R}$, in other hand $\mathbf{\Sigma \gamma} = \lambda \mathbf{\gamma} \implies \mathbf{\bar{\gamma}'\Sigma \gamma} = \mathbf{\bar{\gamma}'\lambda \gamma} = \lambda \mathbf{\gamma'\gamma} \in \mathbb{R}$ then $\lambda \in \mathbb{R}$. Hence $\mathbf{\Lambda}$ it's a diagonal matrix formed by the eigenvalues (orderetd reals) of $\mathbf{\Sigma}$.

Lets $\mathbf{\gamma_1, \gamma_2}$ eigenvectors of $\mathbf{\Sigma}$ with its corresponding eigenvalues $\lambda_1,\lambda_2$ then:

$$
\begin{align*}
\lambda_1<\mathbf{\gamma_1, \gamma_2}>&=\lambda_1\mathbf{\gamma_1'\gamma_2}\\
&=\mathbf{(\lambda_1\gamma_1)'\gamma_2}\\
&=\mathbf{(\Sigma \gamma_1)'\gamma_2}\\
&=\mathbf{\gamma_1'\Sigma'\gamma_2}\\
&=\mathbf{\gamma_1'\Sigma \gamma_2}\\
&=\mathbf{\gamma_1'(\Sigma \gamma_2)}\\
&=\mathbf{\gamma_1'(}\lambda_2 \mathbf{\gamma_2)}\\
&=\lambda_2 \mathbf{\gamma_1' \gamma_2}\\
\lambda_1<\mathbf{\gamma_1, \gamma_2}>&=\lambda_2<\mathbf{\gamma_1, \gamma_2}>\\
\lambda_1<\mathbf{\gamma_1, \gamma_2}>-\lambda_2<\mathbf{\gamma_1, \gamma_2}>&=0\\
(\lambda_1-\lambda_2)<\mathbf{\gamma_1, \gamma_2}>&=0\\
\end{align*}
$$

Hence $\lambda_1 \neq \lambda_2 \implies (\lambda_1 - \lambda_2) \neq 0 \implies <\mathbf{\gamma_1, \gamma_2}>=0$ then $\mathbf{\gamma_1, \gamma_2}$ are ortogonals and $\mathbf{\Gamma}$ it's an ortogonal matrix formed by the ordered eigenvectors of $\mathbf{\Sigma}$.

**Note**: This result works for every covariance matrix included the sample covariance matrix S, Moreover, it works for any definite semipositive definite matrix


**Population overall variability**

At the same case of $\mathbf{\mathbf{S}}$ its possible define a generalized overall varibility measures in terms of $\mathbf{\Sigma}$, this is:

* **Generalized pupulation variance**: $\left|\mathbf{\Sigma}\right| = \left|\mathbf{\Gamma \Lambda \Gamma'}\right| = \mathbf{\left|\Gamma\right|\left|\Lambda\right|\left|\Gamma^{-1}\right|}=\left|\Lambda\right|=(\lambda_1)\cdots(\lambda_p)$
* **Total population variance:** $tr(\mathbf{\Lambda})= \lambda_1+\cdots+\lambda_p$

###  Constant probability density contours and confidence hyper-ellipses.
___

Since $\mathbf{\Sigma}$ it's semipositive definite, for $\mathbf{e}$ an eigenvector of $\mathbf{\Sigma}$, the following inequality it's satisfy

$$0<\mathbf{\gamma' \Sigma \gamma}= \mathbf{\gamma' \lambda \gamma} = \lambda \mathbf{\gamma'\gamma} = \lambda \mathbf{\left|\left|\gamma\right|\right|^2}=\lambda \implies 0<\lambda$$

In other hand by the fact of $\mathbf{\Sigma}$ can be decomposed through the spectral theorem tis possible obtain $\mathbf{\Sigma}$ in terms of these factorization:

$$
\begin{align*}
\mathbf{\Sigma}^{-1} &= \left( \mathbf{\Gamma \Lambda \Gamma}^{-1} \right)^{-1}\\
&= \mathbf{\Gamma \Lambda}^{-1} \mathbf{\Gamma}'\\
&= \left(\mathbf{\gamma_1,...,\gamma_p}\right) diag\left(\lambda_1^{-1}|...,\lambda_p^{-1}\right)\left(\mathbf{\gamma_1,...,\gamma_p}\right)'\\
&= \left(\lambda_1^{-1}\mathbf{\gamma_1} ,..., \lambda_p^{-1}\mathbf{\gamma_p}\right) \left(\mathbf{\gamma_1',...,\gamma_p'}\right)\\
&= \left( \begin{array}{ccccc}
\lambda_{1}\gamma_{1 1}& \cdots & \lambda_{p}\gamma_{1 p}\\
\vdots  & \ddots & \vdots \\
\lambda_{1} \gamma_{p 1}& \cdots & \lambda_{p} \gamma_{p 1} \end{array} \right)
\left( \begin{array}{ccccc}
\gamma_{1 1}& \cdots &\gamma_{p 1} \\
\vdots  & \ddots & \vdots \\
\gamma_{1 p}& \cdots &\gamma_{p p} \end{array} \right)\\
&= \lambda_{1}
\left( \begin{array}{ccccc}
\gamma_{1 1} \gamma_{1 1}  & \cdots & \gamma_{1 1} \gamma_{1 p} \\
\vdots  & \ddots & \vdots \\
\gamma_{p 1} \gamma_{1 1} & \cdots & \gamma_{p 1} \gamma_{p 1} \end{array} \right)
+\cdots+
\lambda_{p}
\left( \begin{array}{ccccc}
\gamma_{1 p} \gamma_{1 p}  & \cdots & \gamma_{1 p} \gamma_{p p} \\
\vdots  & \ddots & \vdots \\
\gamma_{p p} \gamma_{1 p} & \cdots & \gamma_{p p} \gamma_{p p} \end{array} \right)\\
&= \lambda_{1}\mathbf{\gamma_1\gamma_1}'+\cdots+\lambda_{p}\mathbf{\gamma_p\gamma_p}'\\
\mathbf{\Sigma}^{-1} &=\sum_{k=1}^p \frac{1}{\lambda_k}\mathbf{\gamma_k\gamma_k}'.....(i)\\
\end{align*}
$$


With an analogous process its possible get a similar expression for $\mathbf{\Sigma}$:

$$\Sigma=\left[\sum_{k=1}^p\lambda_k\mathbf{\gamma_k'\gamma_k}\right].....(i-b)$$

The equations $(i)$ and $(i-b)$ shows that $\mathbf{\Sigma}$ and $\mathbf{\Sigma}^{-1}$ can be expressed as a linear combination of outer products $\mathbf{\gamma_k \gamma_k}'$, note that each of this products produces a symmetric matrix, then the sum of all products it's a symmetric matrix then $\mathbf{\Sigma}^{-1}$ it's a symmetric matrix. 

$$
\begin{align*}
(\mathbf{x-\mu})'\mathbf{\Sigma}^{-1}(\mathbf{x-\mu})
&=(\mathbf{x-\mu})'\left[\sum_{k=1}^p \frac{1}{\lambda_k}\mathbf{\gamma_k\gamma_k}'\right](\mathbf{x-\mu})\\
&=\sum_{k=1}^p \left[\frac{1}{\lambda_k}\right][(\mathbf{x-\mu})'\mathbf{\gamma_k}][\mathbf{\gamma_k}'(\mathbf{x-\mu})]\\
&=\sum_{k=1}^p \left[\frac{1}{\lambda_k}\right][(\mathbf{x-\mu})\mathbf{\gamma_k}']'[\mathbf{\gamma_k}'(\mathbf{x-\mu})]\\
&=\sum_{k=1}^p \left[\frac{1}{\lambda_k}\right][\mathbf{\gamma_k}'(\mathbf{x-\mu})]^2\\
\implies 0 \leq (\mathbf{x-\mu})'\mathbf{\Sigma}^{-1}(\mathbf{x-\mu})&=\sum_{k=1}^p \left[ \frac{\mathbf{\gamma_k}'(\mathbf{x-\mu})}{\sqrt{\lambda_k}}\right]^2.....(ii)\\
\end{align*}
$$

From $(ii)$ come off, $\mathbf{\Sigma}^{-1}$ it's semi positive definite. The geometric place given by the level sets of the density function can be obtained equaling teh density function of some contant $c_0$ such that $c_0 \in range_{f_{X_1,...,X_p}(x_1,...,x_p)}$.

$$
\begin{align*}
c_0 &= f_{X_1,...,X_p}(x_1,...,x_p)\\
c_0 &= \frac{\left| \Sigma \right|^{-\frac{1}{2}}}{\left[\sqrt{2 \pi}\right]^p} exp\left( \frac{1}{2} \left[\textbf{x}-\mu \right]' \Sigma^{-1} \left[\textbf{x} - \mu \right] \right) \\
c_0 \left| \Sigma \right|^{\frac{1}{2}} \left[\sqrt{2 \pi}\right]^p  &= exp\left( \frac{1}{2} \left[\textbf{x}-\mu \right]' \Sigma^{-1} \left[\textbf{x} - \mu \right] \right) \\
c_0 \sqrt{\lambda_1\cdots\lambda_p} \left[\sqrt{2 \pi}\right]^p  &= exp\left( \frac{1}{2} \left[\textbf{x}-\mu \right]' \Sigma^{-1} \left[\textbf{x} - \mu \right] \right) \\
\frac{ln(c_0)+ln(\sqrt{\lambda_1\cdots\lambda_p})+ln(\left[\sqrt{2 \pi}\right]^p)}{2}&=\left[\textbf{x}-\mu \right]' \Sigma^{-1} \left[\textbf{x} - \mu \right]\\
c^2&=\left[\textbf{x}-\mu \right]' \Sigma^{-1} \left[\textbf{x} - \mu \right]\\
c^2&=\sum_{k=1}^p \left[ \frac{\mathbf{\gamma_k}'(\mathbf{x-\mu})}{\sqrt{\lambda_k}}\right]^2\\
1&=\sum_{k=1}^p \left[ \frac{\mathbf{\gamma_k}'(\mathbf{x-\mu})}{c\sqrt{\lambda_k}}\right]^2.....(iii)\\
\end{align*}
$$

This equation $(iii)$ represents the standard equation of ellipsoid in $\mathbb{R}^p$ centered at $\mathbf{\mu}$ with p axes which directions are $c\sqrt{\mathbf{\lambda_k}}\gamma_k$. A natural question it's what is the probability that the random vector takes a value inside the hyper ellipsoid?. In general terms the procedure consists in compute an integral in $\mathbb{R}^p$ of the region enclosed within the solid, but it's possible computing this probability without complex integrals just deriving the distribution of $(\mathbf{x-\mu})'\mathbf{\Sigma}^{-1}(\mathbf{x-\mu})$ which is a known classical distribution. 

Since $(\mathbf{x-\mu})'\mathbf{\Sigma}^{-1}(\mathbf{x-\mu})$ it's easy to see that the random vector $\mathbf{x-\mu} \sim N_p(\mathbf{0,\Sigma})$, now consider the j-th term of it's expansion in terms of eigenvalues-eigenvectors. Using the property 1 of the multivariate normal distribution, this vector distributes as follows:

$$
\begin{align*}
\frac{1}{\sqrt{\lambda_j}}\mathbf{\gamma_j}'(\mathbf{x-\mu}) &\sim N\left(0,\left[\frac{1}{\sqrt{\lambda_j}}\mathbf{\gamma_j}'\right]'\Sigma \left[\frac{1}{\sqrt{\lambda_j}}\mathbf{\gamma_j}'\right] \right)\\
\left[\frac{1}{\sqrt{\lambda_j}}\mathbf{\gamma_j}'\right]'\Sigma \left[\frac{1}{\sqrt{\lambda_j}}\mathbf{\gamma_j}'\right]
&= \left[\frac{1}{\sqrt{\lambda_j}}\mathbf{\gamma_j}\right]
\Sigma \left[\frac{1}{\sqrt{\lambda_j}}\mathbf{\gamma_j}'\right] \\
&= \left[\frac{1}{\sqrt{\lambda_j}}\mathbf{\gamma_j}\right]
\left[\sum_{k=1}^p\lambda_k\mathbf{\gamma_k'\gamma_k}\right]
\left[\frac{1}{\sqrt{\lambda_j}}\mathbf{\gamma_j}'\right] \\
&=\left[\mathbf{\gamma_j}\right]
\left[\sum_{k=1}^p \left(\frac{\lambda_k}{\sqrt{\lambda_j}^2}\right) \mathbf{\gamma_k'\gamma_k}\right]
\left[\mathbf{\gamma_j}'\right] \\
&=\sum_{k=1}^p \left(\frac{\lambda_k}{\lambda_j}\right) \mathbf{\gamma_j}\mathbf{\gamma_j'\gamma_k}\mathbf{\gamma_j}' \quad \gamma_j'\gamma_k=0 \quad i\neq k\\
&=\left(\frac{\lambda_j}{\lambda_j}\right)[\mathbf{\gamma_j}\mathbf{\gamma_j'}][\mathbf{\gamma_j}\mathbf{\gamma_j}'] \\
&=(1) \left[ \left\|\left| \mathbf{\gamma_j} \right|\right|^2 \right] \left[ \left\|\left| \mathbf{\gamma_j}\right|\right|^2 \right]\\
\left[\frac{1}{\sqrt{\lambda_k}}\mathbf{\gamma_k}'\right]'
\sum_{k=1}^p \lambda_k\mathbf{\gamma_k'\gamma_k}
\left[\frac{1}{\sqrt{\lambda_k}}\mathbf{\gamma_k}'\right] &= 1\\
\implies \frac{1}{\sqrt{\lambda_k}}\mathbf{\gamma_k}'(\mathbf{x-\mu}) & \sim N\left(0,1\right)
\end{align*}
$$

The above result facilitates the probability computing of the next event:

$$
\begin{align*}
P\left[ (\mathbf{x-\mu})'\mathbf{\Sigma}^{-1}(\mathbf{x-\mu}) \leq \mathcal{c} \right]
&= P\left[\sum_{k=1}^p \left[ \frac{\mathbf{\gamma_k}'(\mathbf{x-\mu})}{\sqrt{\lambda_k}}\right]^2 \leq \mathcal{c} \right]\\
&= P\left[\sum_{k=1}^p \left[Z_k\right]^2 \leq \mathcal{c} \right] \qquad Z_k \sim N(0,1)\\
&= P\left[\mathit{W}^2 \leq \mathit{c} \right] \qquad \mathit{W}^2 \sim \chi^2(p)\\
P\left[ (\mathbf{x-\mu})'\mathbf{\Sigma}^{-1}(\mathbf{x-\mu}) \leq \mathcal{c} \right] &= F_{\chi_p^2}(\mathcal{c})
\end{align*}
$$

Then $\mathcal{c}$ it's a critical value of a chi squared quantile function with $p$ freedom degrees $\chi^2_p(1-\alpha)$ for some $\alpha \in (0,1)$, hence:

$$
\begin{align*}
P\left[ (\mathbf{x-\mu})'\mathbf{\Sigma}^{-1}(\mathbf{x-\mu}) \leq \chi^2_p(1-\alpha) \right] = 1-\alpha\\
P\left[ 1=\sum_{k=1}^p \left[ \frac{\mathbf{\gamma_k}'(\mathbf{x-\mu})}{\chi^2_p(1-\alpha)\sqrt{\lambda_k}}\right]^2 \right] = 1-\alpha\\
\end{align*}
$$

Returning to $(iii)$ and replacing the chi squared critical value $\chi^2_p(1-\alpha)$ in to the hyper-ellipse equation, it's possible construct hyper-ellipses such that it envelopes the region with a confidence of $(1-\alpha)$%. The border of this region it's called **constant probability density contours**

$$1=\sum_{k=1}^p \left[ \frac{\mathbf{\gamma_k}'(\mathbf{x-\mu})}{\chi^2_p(1-\alpha)\sqrt{\lambda_k}}\right]^2$$

# Maximum likelihood estimators
___

Lets the data matrix $\mathbf{X}_{n \times p}$ such that each row of this data set it's independen one each other, the data matrix could be represent independt sample of the muultivariated normal distribution, since the rows are independent the jointly distribution density function could be obtained multiplying the individual densities.

$$
\begin{align*}
f_{\mathbf{x_1,...,x_n}}(X_1,..,X_p)&=\prod_{i=0}^{n} \left[\frac{\left| \Sigma_i \right|^{-\frac{1}{2}}}{\left[\sqrt{2 \pi}\right]^p} exp\left( \frac{1}{2} \left[\textbf{x}_i-\mu \right]' \Sigma^{-1} \left[\textbf{x}_i - \mu\right] \right)\right]\\
&=\left[\frac{\left|\Sigma\right|^{-\frac{1}{2}}}{\left[\sqrt{2\pi}\right]^p}\right]^n \prod_{i=0}^{n}\left[exp\left(\frac{1}{2} \left[\textbf{x}_i-\mu\right]'\Sigma^{-1}\left[\textbf{x}_i-\mu\right]\right)\right]\\
\mathcal{L}\left(\mathbf{\mu,\Sigma}|\mathbf{x_1,...,x_n}\right) &=\frac{\left|\Sigma\right|^{-\frac{n}{2}}}{\left[\sqrt{2\pi}\right]^{np}} exp\left(\sum_{i=1}^n\frac{1}{2}\left[\textbf{x}_i-\mu\right]'\Sigma^{-1}\left[\textbf{x}_i-\mu\right]\right)
\end{align*}
$$

Due to $\mathbf{x_1,...,x_n}$ it's fixed values in the data matrix if $\mu$ and $\mathbf{\Sigma}$ are unknown then the density function could be put as function of the parameters given de data $\mathbf{x_1,...,x_n}$, this function $\mathcal{L}\left(\mathbf{\mu,\Sigma}|\mathbf{x_1,...,x_n}\right)$ is called the likelihood function. The maximum likehood estimators of $\mu$ and $\Sigma$b are the values $\hat{\mu}$ and $\hat{\mathbf{\Sigma}}$ which their maximum of $\mathcal{L}$ is reached. For the multivariate normal distribution the maximum likehood estimators is:

$$\hat{\mu}=\bar{\mathbf{x}} \qquad \hat{\mathbf{\Sigma}}=\mathbf{S}$$

Maximum likehood estimators possess and invariance property. So if $\mathbf{\theta}$ is a unknow parameter and $\hat{\mathbf{\theta}}$ is its ml-estimator and $g$ a function of $\mathbf{\theta}$ then the ml-estimator of $\widehat{g(\mathbf{\theta})}$ is $g(\widehat{\mathbf{\theta}})$, using this propertie the following estimators could be computing as functions of $\mathbf{\bar{x}}$ and $\mathbf{S}$.


* $\widehat{\mathbf{\Sigma}^{-1}}=\mathbf{S}^{-1}$
* $\widehat{\mathbf{\Gamma\Lambda\Gamma^{-1}}}=\mathbf{QLQ^{-1}}$ where $\mathbf{S=QLQ^{-1}}$

**steep 2** Coding functions to perform the constants probability density contours analysis

* **computeMaximumLikelihoodEstimators**: Compute the sample mean vector $\mathbf{\bar{x}}$ and sample covariance matrix $\mathbf{S}$ since these statistics are the maximum likelihood estimators of the parameters of normal multivariate istribution.

* **diagonalize**: Performs the spectral decomposition of the sample covariance matrix, due to the fact of the spectral decomposition it's a function of $\mathbf{\Sigma}$ by the invariance theorem the eigenvalues and eigenvectors geting from decompose the sample covariance matrix will be the estimators of the poblational eigenvectors and eigenvalues.

* **plotConfidenceEllipse**: Once the chi squared critical value is given, it's time to plot the constant probability density contour in direction of the eigenvectors with length eigenvalue times chi squared critical value

* **getProbabilityDensityContour**: Puts in order the last function and perform the analysis.

```python
from pyspark.mllib.linalg.distributed import RowMatrix
from pyspark.ml.linalg import Vectors
from pyspark.mllib.stat import Statistics
from scipy.stats import chi2
from matplotlib.patches import Ellipse

def computeMaximumLikelihoodEstimators(dataSetDF):
    mu = np.array(dataSetDF.groupby().mean().rdd.map(lambda r: r[:]).top(1)[0])
    sigma = RowMatrix(dataSetDF.rdd.map(lambda r: [r[:]])).computeCovariance().toArray()
    return mu, sigma

def diagonalize(sigma):
    eigenValues, eigenVectors = np.linalg.eigh(sigma)
    order = eigenValues.argsort()
    return eigenValues, eigenVectors

def plotConfidenceEllipse(plt,mu,eigenValues,eigenVectors,chiSquaredCriticalVale,ax=None,**kwargs):
    if ax is None:
        ax = plt.gca()
    theta = np.degrees(np.arctan2(*eigenVectors[:,0][::-1]))
    width,height = chiSquaredCriticalVale*np.sqrt(eigenValues)
    confidenceEllipse = Ellipse(xy=mu,width=width,height=height,angle=theta,**kwargs)
    ax.add_artist(confidenceEllipse)
    return confidenceEllipse

def getProbabilityDensityContour(plt,datasetDF,cols,alpha,freedomDegrees,color,name='DataSet'):
    mu, sigma = computeMaximumLikelihoodEstimators(datasetDF.select(cols))
    eigenValues, eigenVectors = diagonalize(sigma)
    chiSquaredCriticalVale = chi2.ppf(q=(1-alpha),df=freedomDegrees)
    plotConfidenceEllipse(plt,mu,eigenValues,eigenVectors,chiSquaredCriticalVale,color=color,alpha=0.25)
    sumaryTable=[]
    sumaryTable.append(['Mean',mu])
    sumaryTable.append(['Covariance matrix',sigma])
    sumaryTable.append(['EigenValues',eigenValues])
    sumaryTable.append(['EigenVectors',eigenVectors])
    sumaryTable.append(['Confidence',1-alpha])
    sumaryTable.append(['Chi-squared critical value',chiSquaredCriticalVale])
    sumaryTableDF = pd.DataFrame(sumaryTable,columns=['Summary',name])
    return sumaryTableDF

def scatterPlot(plt,dataDF,col_1,col_2,color):
    dataPD = dataDF.select([col_1,col_2]).toPandas()
    plt.scatter(dataPD[col_1],dataPD[col_2],color=color)
```

# Simulating multivariate normal random vectors
___

It's possible to generate a random vector that follows a normal multivariate distribution given a fixed covariance matrix $\mathbf{\Sigma}$, mean vector $\mathbf{\mu}$ and a multivariate standarized normal vector $\mathbf{z}$. Retaking from the derivation of the multivariate normal distribution, It had been deduced that the normal multivariate distribution could be written as an affine transformation (a linear transformation followed by a translation) of the standardized multivariate normal vector of dimension p, as indicated below:

$$
\begin{align*}
\left[ \begin{array}{ccccc}
X_{1} \\
\vdots  \\
X_{p} \end{array} \right]
& = 
\left[ \begin{array}{ccccc}
\mu_{1} \\
\vdots  \\
\mu_{p} \end{array} \right]
+
\left[ \begin{array}{ccccc}
a_{1 1} & \cdots & a_{1 p} \\
\vdots  & \ddots & \vdots \\
a_{p 1} & \cdots & a_{p p} \end{array} \right]
\left[ \begin{array}{ccccc}
Z_{1} \\
\vdots  \\
Z_{p} \end{array} \right]\\
\textbf{x} & = \mu +\textbf{Az} \\
\end{align*}
$$

It was known that the matrix $\mathbf{A}$ has to satisfy:

$$
\begin{align*}
\mathbf{AA'}&=\mathbf{\Sigma}\\
&=\mathbf{\Gamma \Lambda \Gamma}^{-1}\\
&=\mathbf{\Gamma \Lambda \Gamma'}\\
&=\mathbf{\left[\Gamma \Lambda^{1/2}\right] \left[\Lambda^{1/2}\Gamma'\right]}\\
\mathbf{AA'}&=\mathbf{\left[\Gamma \Lambda^{1/2} \right] \left[\Gamma \Lambda^{1/2}\right]'}\\
\implies \mathbf{A} &= \mathbf{\Gamma \Lambda^{1/2}}\\
\end{align*}
$$

Where $\mathbf{\Lambda^{1/2}}$ Is the diagonal matrix of square roots of the eigenvalues, so given a multivariate standar normal vector $\mathbf{z}$ the following transformation distributes multivariate normal with parameters $\mathbf{\mu}$ and $\mathbf{\Sigma}$.

$$
\begin{align*}
\textbf{x} & = \mu +\mathbf{\left[\Gamma \Lambda^{1/2} \right] z} \sim N_p(\mathbf{\mu},\mathbf{\Sigma}) \\
\end{align*}
$$

**Example 2.1 - 3 
[Generating a sample of multivariate normal distribution in two dimensions]**

**step 1**: Defining some functions:

* **mvNormalizerMatrix** This function generate de dcomposition of $\mathbf{\Sigma}$ in to the product of matrices $\mathbf{\left[\Gamma \Lambda^{1/2}\right] \left[\Lambda^{1/2}\Gamma'\right]}$ and return just one of them: $\mathbf{\left[\Gamma \Lambda^{1/2}\right]}$

* **multivariateNormalVector**: This function generates a multivariate standar normal vector Z using the method of the quantile function (ppf) to a random number between 0 and 1 uniform distributed for later apply an affine transformation with parameters Shift (the traslation vector) and LinearTransform (The matrix) this transformation produces a multivariate normal vector of dimension p with parameters $Shift$ and $(LinearTransform)(LinearTransform)'$



```python
from numpy import linalg as linear
from numpy import random as ran
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
import math as mt

def mvNormalizerMatrix(Sigma):
    eigenValues, eigenVectors = diagonalize(Sigma)
    LambdaSqrt = np.diag([np.sqrt(v) for v in eigenValues])
    return np.dot(LambdaSqrt,eigenVectors), eigenValues, eigenVectors

def multivariateNormalVector(X,Shift,LinearTransform):
    Z = [one*norm.ppf(ran.uniform()) for one in np.ones(len(Shift))]
    X = [float(x) for x in np.dot(Z,LinearTransform)]
    for i in range(0,len(X)): X[i]=X[i]+Shift[i]
    return X
```

**step 2**: Generating a sample of size n of the multivariate normal distribution:

$$
(X_1,X_2) \sim N\left(
\mu=\left[ \begin{array}{ccccc}
0 \\
0 \end{array} \right]
,
\Sigma=\left[ \begin{array}{ccccc}
5 & 5\\
5 & 10 \end{array} \right]
\right) 
$$

```python
n=5000
mu=[0,0]
sigma = [[5,5],[5,10]]
alpha_0=0.01
degreesFreedom=len(mu)
mormalizerMatrix, eigenValues, eigenVectors=mvNormalizerMatrix(sigma)
```

Applying the transformation in to RDD with spark

```python
normalSampleDF = sc.parallelize(np.ones(n))\
                 .map(lambda y: multivariateNormalVector(y,mu,mormalizerMatrix))\
                 .toDF(['x_1','x_2'])
```

**Step 3** Visualizing the sample and the constant density ellipse for differents leves of confidence $(1-\alpha)$%:

```python
plt.figure(figsize=(15,15))
plt.axis([-12,12,-15,15])
scatterPlot(plt,normalSampleDF,'x_1','x_2','Black')
for i in range(0,6):
    chi2CV = chi2.ppf(q=(1-alpha_0*i),df=freedomDegrees)
    plotConfidenceEllipse(plt,mu,eigenValues,eigenVectors,chi2CV,color='Black',alpha=0.1)
plt.show()
```

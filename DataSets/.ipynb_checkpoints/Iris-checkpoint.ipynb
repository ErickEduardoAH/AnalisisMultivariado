```python
%matplotlib inline
import pandas as pd
import matplotlib
matplotlib.style.use('ggplot')
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
```

```python
def eigsorted(cov):
    vals, vecs = np.linalg.eigh(cov)
    order = vals.argsort()[::-1]
    return vals[order], vecs[:,order]

def plot_point_cov(points, nstd=2, ax=None, **kwargs):
    pos = points.mean(axis=0)
    cov = np.cov(points, rowvar=False)
    return plot_cov_ellipse(cov, pos, nstd, ax, **kwargs)

def plot_cov_ellipse(cov, pos, nstd=2, ax=None, **kwargs):
    if ax is None:
        ax = plt.gca()
    vals, vecs = eigsorted(cov)
    theta = np.degrees(np.arctan2(*vecs[:,0][::-1]))
    width, height = 2 * nstd * np.sqrt(vals)
    #return(vals)
    ellip = Ellipse(xy=pos, width=width, height=height, angle=theta, **kwargs)
    ax.add_artist(ellip)
    return ellip                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                             
```

```python
broken_df = pd.read_csv('../Iris/iris.csv')
setosa_petal = broken_df[broken_df['class'] == 'Iris-setosa'][['petal_length','petal_width']]
versicolor_petal = broken_df[broken_df['class'] == 'Iris-versicolor'][['petal_length','petal_width']]
virginica_petal = broken_df[broken_df['class'] == 'Iris-virginica'][['petal_length','petal_width']]
```

```python
plt.figure(figsize=(20,10))
plt.scatter(setosa_petal['petal_length'],setosa_petal['petal_width'],color='red')
plt.scatter(versicolor_petal['petal_length'],versicolor_petal['petal_width'],color='blue')
plt.scatter(virginica_petal['petal_length'],virginica_petal['petal_width'],color='green')
```

```python
plt.figure(figsize=(20,10))
plt.axis([-1,8,-1,3])

plot_point_cov(np.asarray(setosa_petal), nstd=2.5, alpha=0.5, color='red') 
plt.plot(setosa_petal['petal_length'],setosa_petal['petal_width'],'ro',color='red')

plot_point_cov(np.asarray(versicolor_petal), nstd=2.5, alpha=0.5, color='blue') 
plt.plot(versicolor_petal['petal_length'],versicolor_petal['petal_width'],'ro',color='blue')

plot_point_cov(np.asarray(virginica_petal), nstd=2.5, alpha=0.5, color='green') 
plt.plot(virginica_petal['petal_length'],virginica_petal['petal_width'],'ro',color='green')

plt.show()
```

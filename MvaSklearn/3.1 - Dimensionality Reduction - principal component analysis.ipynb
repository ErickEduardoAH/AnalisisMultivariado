A principal component analysis is concerned with explained the variance  structure of a set of linear combinations whose general objetive its reduce data dimension and interpretation. Algebraically, principal components are a particular linear combination of p random variables $X_1,...,X_p$ that represent a change of basis obtained the new axes by the rotation of the original system. The new axes represent the directions with maximum variability and provides a simple and parsimonious description of the covariance structure.

**Note**: Theire development desnt require a multivariate normal assumption, but principal components derived from multivariate normal populations have a usefull interpretations in terms of constant density ellipsoids. 

Lets the random vector $\mathbf{x}=(X_1,...,X_p) \in \mathbb{R}^p$ with covariance matrix $\mathbf{\Sigma}$ and the asociated ordered eigenvalues $\lambda_1 \geq ... \geq \lambda_p \geq 0$ and consider the following linear combinations given by the row vectors $\mathbf{a_i}$.

$$
\begin{cases}
Y_{1}   & = a_{1 1} X_{1} + \cdots + a_{1 p} X_{p} = \mathbf{a_1 x} \\
\vdots  & = \vdots \\
Y_{p}   & = a_{p 1} X_{1} + \cdots + a_{p p} X_{p} = \mathbf{a_2 x} \\
\end{cases}
\\
\begin{align*}
\left[ \begin{array}{ccccc}
Y_{1} \\
\vdots  \\
Y_{p} \end{array} \right]
& = 
\left[ \begin{array}{ccccc}
a_{1 1} & \cdots & a_{1 p} \\
\vdots  & \ddots & \vdots \\
a_{p 1} & \cdots & a_{p p} \end{array} \right]
\left[ \begin{array}{ccccc}
X_{1} \\
\vdots  \\
X_{p} \end{array} \right]\\
\\
\textbf{y} & =\textbf{Ax} \\
\end{align*}
$$

Using the last results, we have for the ith row $\mathbf{a_i}$ of $\mathbf{A}$: 

$$
V(Y_i) = \mathbf{a_i \Sigma a_i'}\\
Cov(Y_i,Y_j) = \mathbf{a_i \Sigma a_j'}\\
$$

We know that $Cov(\mathbf{y}) = \mathbf{A \Sigma A'}$ but the covariances from this expression could be increased by multiplying any constant by $\mathbf{A}$, to quote this indeterminacy it's necesary to restrict to coeficient vectors of lenght 1. The principal components are those uncorrelated linear combinations $Y_1,...Y_p$ whose variances are larger as possible such that $Y_i$ is independen from $Y_j$ or $i \neq j$.

* First principal component: linear combination $\mathbf{a_1 x}$ that maximizes $V(\mathbf{a_1 x})$ subject to $<\mathbf{a_1,a_1}>$=1
* Second principal component: linear combination $\mathbf{a_1 x}$ that maximizes $V(\mathbf{a_1 x})$ subject to $<\mathbf{a_2,a_2}>$=1 and $<\mathbf{a_1,a_2}>$=0
$$\vdots$$
* The ith principal component = linear combination $\mathbf{a_i x}$ that maximizes $V(\mathbf{a_i x})$ subject to $<\mathbf{a_i,a_i}$>=1 and $<\mathbf{a_i,a_k}>$=0 for $k < i$

Sinces the covariance matrix $\mathbf{\Sigma}$ it's orthogonally diagonalizable, $\mathbf{\Sigma}=\mathbf{\Gamma \Lambda \Gamma'}$, If  $\lambda_k \mathbf{\gamma_k}$ it's a eigen vector asociated to $\mathbf{\Sigma}$, it's satisfy the principal components restrictions:

1. $<\mathbf{\gamma_i',\gamma_i}$>=1
1. $<\mathbf{\gamma_i',\gamma_0}$>=0 for $k < i$

It remains to prove that variance as function of $\mathbf{a}$ it's quoted and the eigenvectors maximize the variance. Considering the restrictions that $\mathbf{a}$ must satify it clear that $\mathbf{a' \Sigma a}$ and $\mathbf{\frac{a' \Sigma a}{a' a}}$ have the same maximum where $\mathbf{\frac{a' \Sigma a}{a' a}}$ eliminates the problem of the constant. Considering the next transformation $\mathbf{b}=\mathbf{\Gamma' a}$ we have:

$$
\begin{align*}
\mathbf{\frac{a' \Sigma a}{a' a}} &= \mathbf{\frac{a' \Gamma \Lambda \Gamma' a}{a' a}}\\
&= \mathbf{\frac{b' \Lambda b}{b' b}} \\
\mathbf{\frac{a' \Sigma a}{aa'}} &= \frac{\sum_{k=1}^{p} \lambda_k b_k^2}{\sum_{k=1}^{p}b_k^2}\\
\end{align*}
$$

Since $\lambda_p \leq \lambda_{p-1} \leq \dots \leq \lambda_i \leq \dots \leq \lambda_1$ for the ordered ith eigenvalue $\lambda_i$ the variance as function of function $\mathbf{a}$ it's quoted.

$$
\begin{cases}
\lambda_i b_1^2 \leq \lambda_i b_1^2  \\
\lambda_{i-1} b_1^2 \leq \lambda_i b_1^2  \\
\vdots  \\
\lambda_p b_p^2 \leq \lambda_1 b_p^2  \\
\end{cases}
$$


$$
\begin{align*}
\lambda_i b_1^2 + \lambda_{i-1} b_2^2 + \dots +\lambda_p b_p^2 &\leq \lambda_i b_1^2 + \lambda_i b_2^2 + \dots +\lambda_i b_p^2\\
\frac{\sum_{k=i}^{p} \lambda_k b_k^2}{\sum_{k=i}^{p}b_k^2} &\leq \frac{\sum_{k=i}^{p} \lambda_i b_k^2}{\sum_{k=i}^{p}b_k^2}\\
\frac{\sum_{k=i}^{p} \lambda_k b_k^2}{\sum_{k=i}^{p}b_k^2} &\leq \lambda_i \frac{\sum_{k=i}^{p} b_k^2}{\sum_{k=i}^{p}b_k^2}\\
\frac{\sum_{k=i}^{p} \lambda_k b_k^2}{\sum_{k=i}^{p}b_k^2} &\leq \lambda_i \\
\end{align*}
$$

In particular for $\lambda_1$ we have that any variance for any linear combinations with coeficients given by $\mathbf{a}$ it's qouted by the first eigenvalue.

$$
\begin{align*}
\frac{\sum_{k=1}^{p} \lambda_k b_k^2}{\sum_{k=1}^{p}b_k^2}&\leq \lambda_1 \\
\mathbf{\frac{a' \Sigma a}{a' a}}&\leq \lambda_1 \\
\end{align*}
$$

If some value of $\mathbf{x}$ the function reach the quote, then this point its a critical point and maximize the function, so replacing the first eigenvalue $\gamma_1$:

$$
\begin{align*}
\mathbf{\frac{a \Sigma a'}{a' a}} &= \mathbf{\frac{\gamma_i' \Sigma \gamma_i}{\gamma_i' \gamma_i}} \\
&= \mathbf{\gamma_i' \Sigma \gamma_i} \\
&= \mathbf{\gamma_i'} \lambda_i \mathbf{\gamma_i} \\
&= \lambda_i \\
\mathbf{\gamma_i' \Sigma \gamma_i} &= \lambda_i \\
V(\mathbf{\gamma_i' x}) &= \lambda_i \\
\end{align*}
$$

If we take $\mathbf{a}=\mathbf{\gamma_i'}$ this transformation reach the maximum variance for $Y_i = \mathbf{\gamma_i' x}$, so $\mathbf{\gamma_i'}$ its the ith principal component, because these are the linear combinations that maximizes the variances and produce new independent variables. 

![pca gif](static/pca2gif.gif)

So the principal components (PCA) transformation of $\mathbf{x}$ with covariance matrix given by $Cov(\mathbf{x})=\mathbf{\Sigma}=\mathbf{\Gamma \Lambda \Gamma'}$  its:


$$
\begin{cases}
\begin{align*}
PCA_1 & = \gamma_{1 1} X_{1} + \cdots + \gamma_{1 p} X_{p} = \mathbf{\gamma_1' x} \\
\vdots & = \vdots \\
PCA_p & = \gamma_{p 1} X_{1} + \cdots + \gamma_{p p} X_{p} = \mathbf{\gamma_p' x} \\
\end{align*}
\end{cases}
$$


$$
\begin{align*}
\begin{bmatrix}
    PCA_1 \\
     \vdots\\
    PCA_p \\
\end{bmatrix}&=
\begin{bmatrix}
    \text{-} & \mathbf{\gamma_1} & \text{-} \\
    & \vdots &\\
    \text{-} & \mathbf{\gamma_p} & \text{-} \\
\end{bmatrix}
\begin{bmatrix}
    X_1 \\
     \vdots\\
    X_p \\
\end{bmatrix}\\
PCA(\mathbf{x}) &= \mathbf{\Gamma' x}  \\
\end{align*}
$$


If we have a data matrix $\mathbf{X}_{n \times p}$ with $\mathbf{S} = \mathbf{\Gamma \Lambda \Gamma'}$ the pc transformations for this matrix it's given by:

$$
PCA(\mathbf{X}_{n \times p}) = \mathbf{X}_{n \times p}\mathbf{\Gamma}
$$

**Explained variability ratios:**

From the spectral analysis we have 

$$
\begin{align*}
\Sigma&=\sum_{k=1}^p\lambda_k\mathbf{\gamma_k\gamma_k'}\\
tr[ \Sigma ] &= tr \left[ \sum_{k=1}^p\lambda_k\mathbf{\gamma_k\gamma_k'} \right]\\
&= \sum_{k=1}^p \lambda_k tr \left[ \mathbf{\gamma_k\gamma_k'} \right]\\
&= \sum_{k=1}^p \lambda_k tr \left[ \mathbf{\gamma_k'\gamma_k} \right]\\
tr[ \Sigma ] &= \sum_{k=1}^p \lambda_k \\
\end{align*}
$$

The percentage of variability that contributes the i-th component is given by:

$$\frac{\lambda_i}{tr[\mathbf{\Sigma}]}=\frac{\lambda_i}{\lambda_1+\dots+\lambda_p}$$

The cumulated percentaje of variability untill the i-th component is:

$$\frac{\sum_{k=1}^{i}\lambda_k}{tr[\mathbf{\Sigma}]}=\frac{\lambda_1+\lambda_i}{\lambda_1+\dots+\lambda_p}$$


The number of principal components is taken so that the accumulated explained variability is close to a value specified by the scientist, for example 80%. On the other hand, if the representation of is stabilized from a certain eigenvalue, then increase the dimension hardly provides more explained variability.



# Biplot
---

A biplot its a generalization of the simple two-variable scatterplot. A biplot allows information on both samples and variables of a data matrix to be displayed graphically. Samples are displayed as points while variables are displayed either as vectors. The prefix "bi" refers to the overlap, in the same plot variables and observtions.

A biplot for a matrix $\mathbf{X_{n \times p}}$ is a graphic representation by p vectors (markers) $v_1,\dots,v_p$ for the n rows $u_1,\dots, u_n$ of $\mathbf{X}$ such that the element $x_{ij} \approx \mathbf{v_i u_j'}$ as well as possible.

$$\mathbf{X \approx UV'}$$

For the euclidean space $\mathbb{R^2}$ if the matrix $\mathbf{X}$ has range 2 the aproximations will be exactly in other case it will be an aproximation. Consider the PCA transformation:

$$
\begin{align*}
\mathbf{\Gamma' x}&=\mathbf{y}\\
\mathbf{x} &=\mathbf{\Gamma y}\\
\end{align*}
\\
\begin{cases}
X_{1}   & = \gamma_{1 1} Y_{1} + \cdots + \gamma_{p 1} Y_{p}\\
\vdots  & = \vdots \\
X_{p}   & = \gamma_{1 p} Y_{1} + \cdots + \gamma_{p p} Y_{p}\\
\end{cases}
$$


Taking the first i ordered components.

$$
\begin{cases}
\begin{align*}
X_{1}   & \approx \gamma_{1 1} Y_{1} + \cdots + \gamma_{i 1} Y_{i}\\
\vdots  & \approx \vdots \\
X_{p}   & \approx \gamma_{1 p} Y_{1} + \cdots + \gamma_{i p} Y_{i}\\
\end{align*}
\end{cases}
$$

The markers will be the first columns of the variables transformed to the new space and the vectors the first columns of the base change matrix. This approximation depends on the number of principal components taken (usually 2), the resolution of the biplot its given by the cumulated percentaje of variability untill the i-th component.

```python
from sklearn.decomposition import PCA

pcaNormalCols = ['pc'+str(i+1) for i in range(0,2)]
print(pcaNormalCols)
pcaNormalSampleModel = PCA(n_components=2).fit(normalSamplePD.values)
pcaNormalSampleMatrix = pd.DataFrame(pcaNormalSampleModel.components_.transpose(),columns=pcaNormalCols)

print('PCA Transformation Matrix:\n')
display(pcaNormalSampleMatrix)
print('PCA Eigenvalues:\n',pcaNormalSampleModel.explained_variance_)

plt.figure(figsize=(5,5))
screePlot(plt,pcaNormalSampleModel.explained_variance_ratio_)
```

Visualizing principal componented of data

```python
def plotProjectedBase(plt,components,projectedCanonicalsPD,colums,axesColor='blue',axesAlpha=1,arrowWidth=1.0,\
                      arrowColor='black',arrowAlpha=1,head_width=0.04,head_length=0.1,shift=2/100):
    M = np.matrix(projectedCanonicalsPD[components].values)
    rows,cols = M.T.shape
    maxes = np.amax(abs(M), axis = 0)
    t = np.linspace(-1000, 1000, 100)
    plt.xlabel(components[0])
    plt.ylabel(components[1])
    for i,l in enumerate(range(0,cols)):
        plt.axes().arrow(0,0,M[i,0],M[i,1],head_width=head_width,head_length=head_length,\
                         color=arrowColor,alpha=arrowAlpha,linewidth=arrowWidth)
        plt.annotate(colums[i],xy=(M[i,0],M[i,1]),xytext=(M[i,0]+shift,M[i,1]+shift),\
                         weight='bold',color=arrowColor)
        plt.plot(t*M[i,0],t*M[i,1],color=axesColor,alpha=axesAlpha)        
    plt.plot(0,0,'ok')
    plt.grid(b=True, which='major')
```

```python
plt.figure(figsize=(15,15))
plt.axis([-10,10,-10,10])
plotProjectedBase(plt,['pc1','pc2'],pcaNormalSampleMatrix,['x1','x2'])
scatterPlot(plt,normalSamplePD,'x1','x2','Black',color='Black')
```

Transforming the data to the new reference system

```python
pcaNormalDataPD = pd.DataFrame(\
                    data=pcaNormalSampleModel.transform(normalSamplePD.values),\
                    columns=pcaNormalCols)

plt.figure(figsize=(15,15))
scatterPlot(plt,normalSamplePD,'x1','x2','Black',color='Black')
scatterPlot(plt,pcaNormalDataPD,'pc1','pc2','Blue',color='Blue')
```

label:

The objective is try to determine if there are significative differences between the observations of red and white wine.

```python
kw=12
labelCol = 'label'
whiteWinnePath = 'DataSets/winequalityWhite.csv'
redWinnePath = 'DataSets/winequalityRed.csv'
```

```python
whiteWinePD = pd.read_csv(whiteWinnePath)
redWinePD = pd.read_csv(redWinnePath)
features = list(redWinePD.columns)
display(whiteWinePD[:3])
display(redWinePD[:3])
```

```python
pcaWineCols = ['pc'+str(i+1) for i in range(0,kw)]
wineX = pd.concat([whiteWinePD,redWinePD]).loc[:,features].values
redWineX = redWinePD.loc[:,features].values
whiteWineX = whiteWinePD.loc[:,features].values
wineX
```

```python
kw=12
centeredWineX = StandardScaler(with_mean=True,with_std=False).fit_transform(wineX)
pcaWineModel = PCA(n_components=kw).fit(centeredWineX)
pcaWineMatrix = pd.DataFrame(pcaWineModel.components_.transpose(),columns=pcaWineCols)
pcaRedWineDataPD = pd.DataFrame(pcaWineModel.transform(redWineX),columns=pcaWineCols)
pcaWhiteWineDataPD = pd.DataFrame(pcaWineModel.transform(whiteWineX),columns=pcaWineCols)
```

```python
print('PCA Transformation Matrix:\n')
display(pcaWineMatrix)
plt.figure(figsize=(5,5))
screePlot(plt,pcaWineModel.explained_variance_ratio_)
```

```python
plt.figure(figsize=(15,15))
scatterPlot(plt,pcaWhiteWineDataPD,'pc1','pc2','Blue',color='Black',alpha=0.5)
scatterPlot(plt,pcaRedWineDataPD,'pc1','pc2','Red',color='Black',alpha=0.5)
plt.show()
```

Something is going wrong with the componentes. A fast inspection reveals that there are some columns with high variance, these columns are absorbing the covariance of the components, this happens because all columns are not in same metric scale, standarizing the columns could be a good transformation in oder to avoid some columns absorbs all the explained variance

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import QuantileTransformer

scaler = RobustScaler().fit(wineX)
scaler = Normalizer().fit(wineX)
#scaler = QuantileTransformer(n_quantiles=10).fit(wineX)
scaler = StandardScaler(with_mean=True,with_std=True).fit(wineX)
```

```python
stdWineX = scaler.transform(wineX)
pcaStdWineModel = PCA(n_components=kw).fit(stdWineX)
pcaStdWineMatrix = pd.DataFrame(pcaStdWineModel.components_.transpose(),columns=pcaWineCols)
print('PCA Transformation Matrix:\n')
display(pcaStdWineMatrix)
plt.figure(figsize=(5,5))
screePlot(plt,pcaStdWineModel.explained_variance_ratio_)
```
